```python
import pandas as pd
import matplotlib.pyplot as plt

# ---------------------------------------------------
# ❶ まず CSV を複数エンコードで読み込む関数
# ---------------------------------------------------
def load_csv_safely(path):
    encodings = ["utf-8-sig", "shift_jis", "cp932", "euc_jp"]
    last_error = None
    for enc in encodings:
        try:
            print(f"Trying encoding: {enc}")
            return pd.read_csv(path, encoding=enc)
        except Exception as e:
            last_error = e
            continue
    raise last_error


# ---------------------------------------------------
# ❷ CSV 読み込み（あなたの CSV ファイル名に変更）
# ---------------------------------------------------
file_path = "jitennsyadeta.csv"   # 例: カレントフォルダに置く
df = load_csv_safely(file_path)

print("読み込み成功!")
print(df.head())

# ---------------------------------------------------
# ❸ 「発生年月日（始期）」 など日付列の名前を自動検出
# ---------------------------------------------------
date_col = None
for col in df.columns:
    if "日" in col or "月" in col or "年" in col:
        date_col = col
        break

if date_col is None:
    raise ValueError("日付らしい列が見つかりません。列名を教えてください。")

print("使用する日付列:", date_col)

# 日付に変換
df["date"] = pd.to_datetime(df[date_col], errors="coerce")
df = df.dropna(subset=["date"])

# ---------------------------------------------------
# ❹ 月ごとの件数カウント
# ---------------------------------------------------
df["month"] = df["date"].dt.month
monthly_counts = df["month"].value_counts().sort_index()

print(monthly_counts)

# ---------------------------------------------------
# ❺ グラフ描画（棒グラフ）
# ---------------------------------------------------
plt.figure(figsize=(10,5))
plt.bar(monthly_counts.index, monthly_counts.values)
plt.xticks(range(1,13))
plt.xlabel("月")
plt.ylabel("事故件数")
plt.title("月ごとの事故件数")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# ====== CSV 読み込み（文字化け対策つき）======
file_path = "jitennsyadeta.csv"  # ←あなたのCSV名に合わせる

# 文字コード自動判定 → ダメなら SJIS・CP932
try:
    df = pd.read_csv(file_path, encoding=None)
except:
    try:
        df = pd.read_csv(file_path, encoding="shift_jis", errors="ignore")
    except:
        df = pd.read_csv(file_path, encoding="cp932", errors="ignore")

# ====== 日付列の読み取り（列名をあなたの CSV に合わせる）======
df["日付"] = pd.to_datetime(df["発生年月日（始期）」"], errors="coerce")

# 月の抽出
df["月"] = df["日付"].dt.month

# ====== 月 → 季節の分類 ======
def to_season(month):
    if month in [12, 1, 2]:
        return "冬"
    elif month in [3, 4, 5]:
        return "春"
    elif month in [6, 7, 8]:
        return "夏"
    else:
        return "秋"

df["季節"] = df["月"].apply(to_season)

# ====== 集計 ======
season_month_count = df.groupby(["季節", "月"]).size().reset_index(name="件数")

# ===== グラフ描画 =====
plt.figure(figsize=(10,6))
plt.bar(season_month_count["月"], season_month_count["件数"])

# ★縦軸・横軸ラベル（あなたのリクエスト部分）★
plt.xlabel("月（1〜12）")     # 横軸
plt.ylabel("事故件数")        # 縦軸

# タイトル
plt.title("季節・月ごとの事故件数")

plt.grid(True)
plt.show()
```
