```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV読み込み
df = pd.read_csv("C:/Users/asami/プログラミング基礎/me/Week6/jitennsyadeta.csv")

# 発生時（始期）を時刻として読み込む
df["発生時"] = pd.to_datetime(df["発生時（始期）"], format="%H:%M", errors="coerce").dt.hour

# 時間帯を分類
def time_zone(h):
    if 0 <= h < 6:
        return "深夜"
    elif 6 <= h < 12:
        return "朝"
    elif 12 <= h < 18:
        return "昼"
    elif 18 <= h <= 23:
        return "夕方"
    else:
        return "不明"

df["時間帯"] = df["発生時"].apply(time_zone)

# グラフ表示
plt.figure(figsize=(12,5))
df["時間帯"].value_counts().plot(kind="bar")
plt.title("時間帯ごとの件数")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 不明を NaN に変換（カラム名を正しく指定）
df["発生時"] = pd.to_numeric(df["発生時"], errors="coerce")

# 時間帯分類の関数
def time_zone(x):
    if pd.isna(x):
        return "不明"
    elif 0 <= x <= 5:
        return "深夜"
    elif 6 <= x <= 11:
        return "朝"
    elif 12 <= x <= 17:
        return "昼"
    elif 18 <= x <= 23:
        return "夕方"
    else:
        return "不明"

# 時間帯列を追加
df["時間帯"] = df["発生時"].apply(time_zone)

# グラフ作成
plt.figure(figsize=(10,5))
df["時間帯"].value_counts().plot(kind="bar")
plt.title("時間帯ごとの件数")
plt.show()
```

```python
import pandas as pd

# ★ ここをあなたの PC のファイルの場所に書き換えてください
file_path = r"C:\Users\asami\プログラミング基礎\me\Week6\jitennsyadeta.csv"

# --- エンコーディング自動判定付きの安全ロード ---
encodings = ["utf-8", "shift_jis", "cp932"]

for enc in encodings:
    try:
        print(f"Trying encoding: {enc}")
        df = pd.read_csv(file_path, encoding=enc, on_bad_lines="skip")
        print("Success!")
        break
    except Exception as e:
        print(f"Failed: {enc}")

# 日付を datetime に変換（文字化け対策）
if "日付" in df.columns:
    df["日付"] = pd.to_datetime(df["日付"], errors="coerce")
elif "発生年月日（始期）" in df.columns:
    df["日付"] = pd.to_datetime(df["発生年月日（始期）"], errors="coerce")

df.head()
```
